```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('img1.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
#convert from BGR to RGB for proper colour displays 
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# plt.imshow('Grayscale Image', gray_image)
plt.imshow(gray_image)
plt.title("Gray Image")
plt.show()
```

```python
plt.imshow(gray_image, cmap='gray')   
plt.show()
```

```python
# Binary Image
_, binary_image = cv2.threshold(gray_image, 128, 255, 50)
plt.imshow(binary_image)
plt.title("Binary_image")
plt.show()
```

```python
img=cv2.imread('img1.jpg')
b, g, r = cv2.split(img)

# Display the channels using matplotlib
plt.figure(figsize=(12, 4))

# Blue channel
plt.subplot(1, 3, 1)
plt.imshow(b, cmap='Blues')
plt.title('Blue Channel')
plt.axis('off')

# Green channel
plt.subplot(1, 3, 2)
plt.imshow(g, cmap='Greens')
plt.title('Green Channel')
plt.axis('off')

# Red channel
plt.subplot(1, 3, 3)
plt.imshow(r, cmap='Reds')
plt.title('Red Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Gaussian Blur
gaussian_blur=cv2.GaussianBlur(img_rgb,(15,15),0)
plt.imshow(gaussian_blur)
plt.show()
```

```python
gaussian_blur=cv2.GaussianBlur(img_rgb,(25,25),0)
plt.imshow(gaussian_blur)
plt.title("Gaussian_Blur")
plt.show()
```

```python
# Median blur
median_blur = cv2.medianBlur(img_rgb, 15)
plt.title('Median Blur')
plt.imshow(median_blur)
plt.show()
 
```

```python
#apply sharpening filter
kernel_sharpening=np.array([[0,-1,0],
                            [-1,5,-1],
                            [0,-1,0]])
sharpened=cv2.filter2D(img_rgb,-1,kernel_sharpening)
```

```python
plt.imshow(sharpened)
plt.show()
```

```python
# apply median filter
median_blur=cv2.medianBlur(img_rgb, 5)
```

```python
plt.imshow(median_blur)
plt.show()
```

```python
# Average filter
average_blur = cv2.blur(img_rgb, (15, 15))
plt.imshow( average_blur)
plt.title('Average Blur')
plt.show()
```

```python
# Apply a Box Filter
kernel_size = (5, 5)  
box_filtered = cv2.blur(img_rgb, kernel_size)
```

```python
plt.imshow(box_filtered)
plt.show()
plt.title('Box filtered')
```
